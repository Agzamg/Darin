```python
import matplotlib.pyplot as plt
import scipy.misc
import numpy as np
import os
import scipy
import imageio
```

```python
f=open('train-1.renju', "r")
data = []
i = 0
for line in f.readlines():
    i += 1
    parsed = line.split(' ')
    data.append(parsed)
```

```python
data = data[:1000]
```

```python
def motiontonumber(motion):
    letter = ord(motion[0]) - ord('a')
    if (letter > 8):
        return (letter * 15 - 15 + int(motion[1:]))
    print(motion[1:])
    return (letter * 15 + int(motion[1:]))
```

```python
def convert(data):
    newdata = []
    j = 0
    for play in data:
        i = 0
        sz = len(play)
        newplay = []
        print(play)
        if (play[0] == "white"):
            newplay.append(0)
        elif (play[0] == "black"):
            newplay.append(1)
        else:
            continue
        for i in range(1, sz):
            newplay.append(motiontonumber(play[i]))
        newdata.append(newplay)
    return newdata
```

```python
newdata = convert(data)
```

```python
X_train = np.zeros((len(newdata), 15, 15, 1))
Y_train = np.zeros(len(newdata))
t = 0
```

```python
def createImagesFromPlay(newdata, k, folder, t):
    i = 0
    fname = folder + '/' + str(i) + '.png'
    initial = np.zeros((15, 15), dtype=np.uint8)
    for move in newdata[k]:
        if (i == 0):
            i += 1
            continue
        fname = folder + '/' + str(i - 1) + '.png'
        if (t < len(newdata)):
            reshaped = np.zeros((15, 15, 1), dtype=np.uint8)
            reshaped[:, :, 0] = initial[:, :]
            X_train[t] = reshaped
            Y_train[t] = move
            t += 1
        initial[move // 15][move % 15] = 128 + 127*(i%2)
        i += 1
        imageio.imsave(fname, initial)
    return t
```

```python
def createAllImages(newdata, t):
    for i in range(len(newdata)):
        folder = str(i)
        print(i)
        if not os.path.exists('./' + folder):
            os.makedirs('./' + folder)
        t = createImagesFromPlay(newdata, i, ('./' + folder), t)
```

```python
createAllImages(newdata, 0)
```

```python
np.save('X_train.npy', X_train)
np.save('Y_train.npy', Y_train)
```

```python
Y_train_vec = np.zeros((len(X_train), 225))
i = 0
for y in Y_train:
    Y_train_vec[i][int(y)] = 1
    i += 1
```

```python
X_train = X_train/256
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D, LeakyReLU, Activation
from keras.layers.normalization import BatchNormalization
from keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Conv2D(32, (2, 2), padding='same', input_shape=(15, 15, 1)))
model.add(LeakyReLU())
model.add(Conv2D(64, (2, 2)))
model.add(LeakyReLU())
model.add(BatchNormalization(axis=3))
model.add(Conv2D(128, (5, 5), padding='same'))
model.add(LeakyReLU())
model.add(Conv2D(128, (5, 5)))
model.add(LeakyReLU())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization(axis=3))
model.add(Flatten())
model.add(Dense(600))
model.add(LeakyReLU())
model.add(Dense(225))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, Y_train_vec, epochs=2, batch_size=64)
```
